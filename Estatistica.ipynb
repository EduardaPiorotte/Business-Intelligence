```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import numpy as np
A= np.array([120, 135, 128, 142, 119, 130, 125, 138, 122, 131])
B= np.array([115, 160, 118, 175, 112, 158, 120, 170, 113, 165])
```

```python
def calcular_Estatistica(dados):
    media = np.mean(dados)
    mediana = np.median(dados)
    variancia = np.var(dados, ddof=1)  # amostral
    desvio_padrao = np.std(dados, ddof=1)
    minimo = np.min(dados)
    maximo = np.max(dados)
    amplitude = maximo - minimo
    cv = (desvio_padrao / media) * 100

    return {
        "Média": media,
        "Mediana": mediana,
        "Variância": variancia,
        "Desvio Padrão": desvio_padrao,
        "Mínimo": minimo,
        "Máximo": maximo,
        "Amplitude": amplitude,
        "Coeficiente de Variacao (%)": cv
    }
```

```python

   estat_A = calcular_Estatistica(A)
estat_B = calcular_Estatistica(B)

print("Servidor A:")
for k, v in estat_A.items():
    print(f"{k}: {v:.2f}")

print("\nServidor B:")
for k, v in estat_B.items():
    print(f"{k}: {v:.2f}")
```

```python
plt.boxplot([A, B], labels=["Servidor A", "Servidor B"])
plt.title("Comparação dos Servidores")
plt.ylabel("Tempo de Resposta (ms)")
plt.show()
```
